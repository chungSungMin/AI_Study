# 수치형 데이터 

```python
import seaborn as sns


data = sns.load_dataset('titanic')
data.head()
```

```python
sns.histplot(data = data, x = 'age')
```

### bins를 적용하게 되면, 막대 그래프의 갯수를 지정할 수 있다. 예를들어 bins = 30 이면 막대그래프 30개로 분포를 표현하게 됩니다

```python
sns.histplot(data = data, x = 'age', bins = 20)
```

### Hue를 사용하게 되면 특정 조건내에서 데이터를 시각화 할 수 있습니다.

```python
sns.histplot(data = data, x = 'age', hue = 'alive')
```

#### multiple을 사용해서 서로 다르게 분리된 그래프에 대해서 어떻게 시각화 할지를 결정할 수 있습니다.

```python
sns.histplot(data = data, x = 'age', hue = 'alive', multiple='dodge')
```

### 주로 sns.dispot()을 활용하여 위에서 그린 histplot()과 kdeplot()을 모두 그릴 수 있다고 합니다.

```python
sns.displot(data = data, x = 'age', kde = True)
```

# 데이터의 관계를 파악하기 위한 시각화

```python
filghts = sns.load_dataset('flights')

filghts.head()
```

```python
filghts_pivot = filghts.pivot(index = 'month',
                              columns= 'year',
                              values='passengers')
```

```python
filghts_pivot
```

```python
sns.heatmap(data=filghts_pivot)
```
