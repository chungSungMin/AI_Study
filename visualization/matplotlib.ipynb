#### matplotlib.pyplot 라이브러리를 사용해서 다양한 데이터들을 시각화 할 수 있습니다.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(1, 2 * np.pi)
y = np.sin(x)
plt.plot(y)
plt.show()
```

```python
x = np.linspace(1, 2 * np.pi)
y = np.sin(x)
# x값의 범위를 제한합니다.
plt.xlim([0,30])
# y 값의 범위를 제한합니다.
plt.ylim([-1, 0.5])
# 그래프의 제목을 설정합니다.
plt.title('limited sin graph')
# x label의 이름을 설정합니다.
plt.xlabel('axis = x')
# y label의 이름을 설정합니다.
plt.ylabel('axis = y')
# 그디드를 설정합니다.
plt.grid(True)
# 내가 원하는 눈금으로 설정가능합니다 ( position위치에 labels를 대응하여 넣어줍니다.)
position = [0, 5, 10, 15, 20, 25, 30]
labels = ['A','B','C','D','E','F','G']
plt.xticks(position, labels)
# 색상 조정이 가능합니다 
plt.plot(y, color = 'r')
plt.show()
```

```python
x = np.linspace(1, 2 * np.pi)
y = np.sin(x)
y2 = np.cos(x)
# x값의 범위를 제한합니다.
plt.xlim([0,30])
# y 값의 범위를 제한합니다.
plt.ylim([-1, 0.5])
# 그래프의 제목을 설정합니다.
plt.title('limited sin graph')
# x label의 이름을 설정합니다.
plt.xlabel('axis = x')
# y label의 이름을 설정합니다.
plt.ylabel('axis = y')
# 그디드를 설정합니다.
plt.grid(True)
# 내가 원하는 눈금으로 설정가능합니다 ( position위치에 labels를 대응하여 넣어줍니다.)
position = [0, 5, 10, 15, 20, 25, 30]
labels = ['A','B','C','D','E','F','G']
plt.xticks(position, labels)
# 색상 조정이 가능합니다 
plt.plot(y, color = 'r')
plt.plot(y2, color ='b')
#legend를 들아서 각 계열의 라벨을 설정 가능합니다.
plt.legend(['sinx', 'cosx'])
plt.show()
```

#### subplot 만들기 

```python
x = np.linspace(1, 2 * np.pi)
y = np.sin(x)

fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(2,3,5)
ax.plot(y)


axi = []
for i in range(6):
    if i == 4 :
        continue
    fig.add_subplot(2, 3, i + 1)
plt.show()
```

```python
fig, ax = plt.subplots(2, 3)
x = np.arange(1,10)
y = np.arange(10,19)

# ax를 flatten()하는것이 중요하다 왜냐하면 1행 2열 이렇게 설정되어있기에 단순히 6개의 연속된 index로 표현하기 위해서 
ax = ax.flatten()

for i in range(6):
    ax[i].plot(x, y)
    ax[i].set_title(f'{i} index plot')
fig.show()
```
