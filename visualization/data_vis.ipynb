# Data 시각화 

#### 꺾은선 그래프 : 데이터의 추세를 파악하는데 유리하다.
#### 막대 그래프 : 두개 이상의 항목을 비교하는데 유리하다.
#### 히스토그램 : 하나의 항목에서 도수를 시각화 하는데 유리하다.
#### 산포도 : 좌표축에서 데이터의 분포를 시각화 하는데 유리하다.
#### 원그래프 : 비율에 따라 데이터를 시각화 하는데 유리하다.

#### *난수* : 컴퓨터는 seed를 기반으로 난수를 생성하게 됩니다. 일반적으로 seed의 경우 컴퓨터의 현재 시간을 기준으로 생성되게 됩니다. 이에 항상 동일한 난수를 생성하고 싶은 경우 seed를 고정 시켜야 합니다. 

```python
import numpy as np

print(f'난수를 설정하지 않음 : {np.random.rand(5)}')
print(f'난수를 설정하지 않음 : {np.random.rand(5)}')

np.random.seed(42)
print(f'난수를 42로 설정함 : {np.random.rand(5)}')
np.random.seed(42)
print(f'난수를 42로 설정함 : {np.random.rand(5)}')

```

#### np.random.rand()의 경우에는 Uniform distribution을 따릅니다.

```python
import matplotlib.pyplot as plt

data = np.random.rand(10000)
plt.hist(data, bins = 'auto')
plt.show()
```

#### np.random.randn()의 경우 정규분포를 따라는 난수를 생성하게 됩니다.

```python
import matplotlib.pyplot as plt

data = np.random.randn(10000)
plt.hist(data, bins = 'auto')
plt.show()
```

#### np.random.binomial(n, p) 의 경우 실패와 성공 2가지로 분류하게 됩니다. p 이상이면 성공 p 미만이면 실패를 출력하게 됩니다.

```python
# 성공확률 0.3으로 100번 시도했을때 성공 횟구를 구하는 실험을 10000번 진행
data = np.random.binomial(100, 0.3, size = 10000)
print(f'각 실험에서 성공이 나온 횟수를 출력합니다 : {data}')
```

#### 리스트에서 랜덤 숫자 N개를 뽑는다.

```python
data = np.array([1,2,3,4,5,6])
# data에서 무작위로 5개의 데이터를 뽑는다 ( 복원 추출 )
print(np.random.choice(data, 5))
```
