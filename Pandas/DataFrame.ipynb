#### lamda 함수 
lambda 매계변수 : 표현식 

example)

def sum(x, y) :
    return x + y
sum(1,3)

위의 코드와 동일하게 lambda 함수를 작성한다면 아래와 같다.
(lambda x, y : x + y)(1,3)

lambda 함수는 매우 간단한 함수 ( 일회성의 함수 ) 를 매우 쉽게 사용할 수 있게 해준다.

#### lambda 함수 배운 김에 간단하게 map()함수도 집고 넘어가기

```python
def sm(x,y):
    return x + y
print(sm(1,3))

print((lambda x, y : x + y)(1,3))


# map의 경우에는 map(함수 , 리스트) 를 작성해서 리스트들의 각 element들을 모두 함수에 적용가능하도록 해줍니다.
lst = [1,2,3,4,5]
result = map((lambda x : x**2), lst)
# 이렇게 하면 map이 iteration 객체를 반환하기에 제대로 확인이 불가능하다.
print(result)
# iteration 객체를 List()를 통해서 풀어준다.
print(list(result))
```

#### 실수값을 카테고리 형식으로 변경하고 싶은 경우

```python
ages = [0, 2, 10, 21, 23, 37, 31, 61, 20, 41, 32, 101]

bins = [1, 20, 30, 50, 70, 100]
labels = ["미성년자", "청년", "중년", "장년", "노년"]
cats = pd.cut(ages, bins, labels=labels)
cats
```

```python
df4 = pd.DataFrame(ages, columns=["ages"])
df4["age_cat"] = pd.cut(df4.ages, bins, labels=labels)
df4
```
