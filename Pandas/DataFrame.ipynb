#### lamda 함수 
lambda 매계변수 : 표현식 

example)

def sum(x, y) :
    return x + y
sum(1,3)

위의 코드와 동일하게 lambda 함수를 작성한다면 아래와 같다.
(lambda x, y : x + y)(1,3)

lambda 함수는 매우 간단한 함수 ( 일회성의 함수 ) 를 매우 쉽게 사용할 수 있게 해준다.

#### lambda 함수 배운 김에 간단하게 map()함수도 집고 넘어가기

```python
def sm(x,y):
    return x + y
print(sm(1,3))

print((lambda x, y : x + y)(1,3))


# map의 경우에는 map(함수 , 리스트) 를 작성해서 리스트들의 각 element들을 모두 함수에 적용가능하도록 해줍니다.
lst = [1,2,3,4,5]
result = map((lambda x : x**2), lst)
# 이렇게 하면 map이 iteration 객체를 반환하기에 제대로 확인이 불가능하다.
print(result)
# iteration 객체를 List()를 통해서 풀어준다.
print(list(result))
```

```python
df1 = pd.DataFrame({
    '고객번호': [1001, 1002, 1003, 1004, 1005, 1006, 1007],
    '이름': ['둘리', '도우너', '또치', '길동', '희동', '마이콜', '영희']
}, columns=['고객번호', '이름'])
df1
```

```python
df2 = pd.DataFrame({
    '고객번호': [1001, 1001, 1005, 1006, 1008, 1001],
    '금액': [10000, 20000, 15000, 5000, 100000, 30000]
}, columns=['고객번호', '금액'])
df2
```

```python
pd.merge(df1, df2)
```

#### outer 방식으로 merge를 진행하는 경우 모든 값들이 표시된다.

```python
pd.merge(df1, df2, how = 'outer')
```

#### left를 기준으로 merge를 진행하였기에, df1의 모든 값들이 들어있다.

```python
pd.merge(df1, df2, how = 'left')
```

#### right을 기준으로 merge를 진행하였기에 df2의 모든 값들이 들어잇다.

```python
pd.merge(df1, df2, how = 'right')
```

```python
df1 = pd.DataFrame({
    '고객명': ['춘향', '춘향', '몽룡'],
    '날짜': ['2018-01-01', '2018-01-02', '2018-01-01'],
    '데이터': ['20000', '30000', '100000']})
df1
```

```python
df2 = pd.DataFrame({
    '고객명': ['춘향', '몽룡'],
    '데이터': ['여자', '남자']})
df2
```

#### df1, df2 둘다 data라는 feature가 존재하는데, 둘다 동일한것을 의미하는게 아니기 떄문에 'on' 함수를 사용해서 기준이되는 feature을 생성한다.

```python
pd.merge(df1, df2,on = '고객명')
```
